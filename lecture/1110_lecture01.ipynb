```python
import googlemaps
```

```python
my_key = "< --본인 API Key 값 입력 --->"
```

```python
maps = googlemaps.Client(key=my_key)
```

```python
place = "서울시립대학교 법학관"
```

```python
geo_location = maps.geocode(place)[0].get('geometry')
```

```python
print(geo_location) 
print("\n")
```

```python
lat = geo_location['location']['lat']
lng = geo_location['location']['lng']
```

```python
print("위도:", lat)
print("경도:", lng)
```
